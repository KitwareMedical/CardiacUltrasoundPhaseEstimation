## Utility Functions

```python
def cvShowImage(imDisp, strName, strAnnotation = '', textColor=(255,255,255)):
    imDisp = imDisp.copy()
    
    if size(strAnnotation) > 0:
        cv2.putText(imDisp, strAnnotation, (10, 20), cv2.FONT_HERSHEY_PLAIN, 2.0, textColor)
        
    cv2.imshow(strName, imDisp)   
    
def mplotShowImage(imInput):
    plt.imshow(imInput, cmap=plt.cm.gray)
    plt.xticks(())
    plt.yticks(())
    
def normalizeArray(a):
    return (0.0 + a - a.min()) / (a.max() - a.min())    

def cvShowVideo(imVideo, strWindowName, waitTime=30, resizeAmount=None):
    
    if not isinstance(imVideo, list):
        imVideo = [imVideo]
        strWindowName = [strWindowName]
        
    blnStop = False      
    while True:
        for fid in range(imVideo[0].shape[2]):
            for vid in range(len(imVideo)):
                
                imCur = imVideo[vid][:,:,fid].astype('uint8')
                
                # resize image if requested
                if resizeAmount:
                    imCur = scipy.misc.imresize(imCur, resizeAmount)
                
                # show image
                cvShowImage(imCur, strWindowName[vid], '%d' % (fid+1))

            # look for "esc" key
            k = cv2.waitKey(waitTime) & 0xff

            if k == 27:
                blnStop = True
                break
        if blnStop:
            break         
    cv2.destroyWindow(strWindowName[vid])
    
def normalizeArray(a):
    return (0.0 + a - a.min()) / (a.max() - a.min())    
    
def writeVideoToFile(imVideo, filename, fourcc=-1, fps=30):

    # start timer
    tStart = time.time()
    
    # write video
    height, width, numFrames = imVideo.shape
    writer = cv2.VideoWriter(filename, fourcc, fps=fps, frameSize=(width, height), isColor=0)

    print writer.isOpened()
    for fid in range(numFrames):
        writer.write(imVideo[:,:,fid].astype('uint8'))
        
    # end timer        
    tEnd = time.time()    
    print 'Writing video took %g {} seconds'.format(tEnd - tStart) 
    
    # release
    writer.release()  
    
def convertFromRFtoBMode( imInputRF ):
    return np.abs(scipy.signal.hilbert(imInputRF, axis=0))
```

## Load Image Data

```python
dataRootDir = '../../data/StephenHeart2015-01-21/'

dataFiles = {'prefix': [],
             'rf': [], 
             'bm': [],
             'bmsc': []}

i = 0
for f in os.listdir(dataRootDir):
    m = re.match('(?P<prefix>.*)_raw.mha$',f) 
    if m:
        print '%d: %s' % (i, m.group('prefix'))
        dataFiles['prefix'].append(m.group('prefix'))
        dataFiles['rf'].append( os.path.join(dataRootDir, f) )
        dataFiles['bm'].append( os.path.join(dataRootDir, "%s_bmode.mha" % m.group('prefix')) )
        dataFiles['bmsc'].append( os.path.join(dataRootDir, "%s_bmode_scanconverted.mha" % m.group('prefix')) )
        i += 1
```

```python
fid = 2
dataFilePath_rf = dataFiles['rf'][fid]
dataFilePath_bm = dataFiles['bm'][fid]
dataFilePath_bmsc = dataFiles['bmsc'][fid]

# Load RF image
print "\nLoading RF image: "
imInput_rf_sitk = sitk.ReadImage(dataFilePath_rf)
print '\tType: ', imInput_rf_sitk.GetPixelIDTypeAsString()
print '\tSize: ', imInput_rf_sitk.GetSize()
print '\tSpacing: ', imInput_rf_sitk.GetSpacing()

imInput_rf = np.transpose( sitk.GetArrayFromImage( imInput_rf_sitk ), [2, 1, 0] )

# Load bmode image before scan conversion
print "\nLoading bmode image before scan conversion: "
imInput_bm_sitk = sitk.ReadImage(dataFilePath_bm)
print '\tType: ', imInput_bm_sitk.GetPixelIDTypeAsString()
print '\tSize: ', imInput_bm_sitk.GetSize()
print '\tSpacing: ', imInput_bm_sitk.GetSpacing()

imInput_bm = np.transpose( sitk.GetArrayFromImage( imInput_bm_sitk ), [2, 1, 0] )
#imInput_bm = convertFromRFtoBMode(imInput_rf)

# Load bmode image before scan conversion
print "\nLoading bmode image after scan conversion: "
imInput_bmsc_sitk = sitk.ReadImage(dataFilePath_bmsc)
print '\tType: ', imInput_bmsc_sitk.GetPixelIDTypeAsString()
print '\tSize: ', imInput_bmsc_sitk.GetSize()
print '\tSpacing: ', imInput_bmsc_sitk.GetSpacing()

imInput_bmsc = np.transpose( sitk.GetArrayFromImage( imInput_bmsc_sitk ), [1, 2, 0] )
```

## Perform Eigen Analysis On Select Scan Lines

```python
# set parameters
numEigenVectors = 10
showEigenVarianceCoverage = False
titlesize = 24
labelsize = 20

thetaSel = np.round(imInput_bm.shape[1] * 0.55)
thetaWindow = 5
minDepth = 0

# display theta line on MIP
plt.figure()

plt.imshow(imInput_bm.max(2).T)

plt.plot([0,imInput_bm.shape[0]], [thetaSel, thetaSel], 'r', linewidth=2)
plt.text(0.05 * imInput_bm.shape[0], thetaSel, 'Angle = %.3d' % (thetaSel+1), 
         fontsize=labelsize, color='r')

plt.plot([minDepth,minDepth], [0, imInput_bm.shape[1]], 'm', linewidth=2)

plt.xlim([0,imInput_bm.shape[0]])
plt.ylim([0,imInput_bm.shape[1]])
plt.xlabel('Depth', fontsize=labelsize)
plt.ylabel('Angle', fontsize=labelsize)
plt.title('Maximum intensity projection of b-mode stack along time', fontsize=20)
plt.gca().invert_yaxis()

# display theta line in the "theta x depth" image at a select time point
timePoint = 10

plt.figure()
plt.imshow(imInput_bm[:,:,timePoint].T)

plt.plot([0,imInput_bm.shape[0]], [thetaSel, thetaSel], 'r', linewidth=2)
plt.text(0.05 * imInput_bm.shape[0], thetaSel, 'Angle = %.3d' % (thetaSel+1), 
         fontsize=labelsize, color='r')

plt.plot([minDepth,minDepth], [0, imInput_bm.shape[1]], 'm', linewidth=2)

plt.xlim([0,imInput_bm.shape[0]])
plt.ylim([0,imInput_bm.shape[1]])
plt.xlabel('Depth', fontsize=labelsize)
plt.ylabel('Angle', fontsize=labelsize)
plt.title('B-mode image at time point %d' %(timePoint+1), fontsize=20)
plt.gca().invert_yaxis()

# display m-mode data at a select theta
plt.figure()
plt.imshow(imInput_bm[:,thetaSel,:].T)

plt.plot([minDepth,minDepth], [0, imInput_bm.shape[2]], 'm', linewidth=2)

plt.xlim([0,imInput_bm.shape[0]])
plt.ylim([0,imInput_bm.shape[2]])

plt.xlabel('Depth', fontsize=labelsize)
plt.ylabel('Time', fontsize=labelsize)
plt.title('M-mode image for Angle=%d' % (thetaSel+1), fontsize=20)
plt.gca().invert_yaxis()

# create a matrix where each row represents the A-line at one time point 
# each row will be treated as an observation for pca
X = imInput_bm[minDepth:,thetaSel,:].T

# perform pca on X
pca = sklearn.decomposition.PCA(n_components=X.shape[0])
pca.fit(X)

# plot percentage of variance explained
if showEigenVarianceCoverage:
    
    x = range(len(pca.components_))
    cumSumVar = np.cumsum(pca.explained_variance_ratio_)
    
    plt.figure()
    plt.plot(x, cumSumVar)
    plt.yticks(np.arange(0.0,1.01,0.05))
    plt.xlabel('Number of principal components')
    plt.ylabel('Percentage of Variance Explained')       

# show eigen vector image
if False:
    plt.figure()
    plt.imshow(pca.components_)
    plt.title('Image in which each row is an eigen vector ordered in decreasing eigen value order', fontsize=titlesize)
    
# project each observation onto the top few eigen vectors
X_proj = pca.transform(X)

# display eigen vectors and the projections onto them
plt.figure(figsize=(30, 7.5))
for eid in range(numEigenVectors):
    
    # display current eigen vector
    plt.subplot(numEigenVectors, 2, eid*2+1)
    plt.plot(pca.components_[eid, :], linewidth=4)   
    plt.xlim([0,X.shape[1]])
    plt.ylabel('%d' % (eid+1), fontsize=labelsize)    
    plt.yticks(())

    if eid == 0:
        plt.title('Top %d eigen vectors for data at angle=%d' % (numEigenVectors, thetaSel+1), 
                  fontsize=titlesize)
        
    if eid == numEigenVectors-1:
        plt.xlabel('Depth', fontsize=labelsize)
    
    # display projection
    plt.subplot(numEigenVectors,2,eid*2+2)
    plt.plot(1 + np.arange(X_proj.shape[0]), X_proj[:,eid], linewidth=4)
    plt.xlim([0,X_proj.shape[0]])
    plt.ylabel('%d' % (eid+1), fontsize=labelsize)    
    plt.yticks(())      

    if eid == 0:
        plt.title('Projections onto top %d eigen vectors for data at angle=%d' % (numEigenVectors, thetaSel+1),
                  fontsize=titlesize)
    
    if eid == numEigenVectors-1:
        plt.xlabel('Time', fontsize=labelsize)
```

## Perform Eigen Analysis On Select Scan Lines Regularized Using a Window

```python
# set parameters
showEigenVarianceCoverage = True
titlesize = 24
labelsize = 20

numEigenVectors = 10
thetaSel = np.round(imInput_bm.shape[1] * 0.55)
thetaWindow = 5
minDepth = 0

# display scan line on b-mode MIP
plt.figure()

plt.imshow(imInput_bm.max(2).T)

plt.plot([0,imInput_bm.shape[0]], [thetaSel, thetaSel], 'r')
plt.text(0.05 * imInput_bm.shape[0], thetaSel, 'Angle = %.3d' % (thetaSel+1), 
         fontsize=labelsize, color='r')
plt.gca().add_patch(mpatches.Rectangle((0, thetaSel-thetaWindow), imInput_bm.shape[0], 2*thetaWindow,
                                       fill=False, ec='g'))

plt.plot([minDepth,minDepth], [0, imInput_bm.shape[1]], 'm', linewidth=2)

plt.xlim([0,imInput_bm.shape[0]])
plt.ylim([0,imInput_bm.shape[1]])
plt.xlabel('Depth', fontsize=labelsize)
plt.ylabel('Angle', fontsize=labelsize)
plt.title('Maximum intensity projection of b-mode stack along time', fontsize=20)
plt.gca().invert_yaxis()

# display scan line on b-mode image at a select time point
timePoint = 10

plt.figure()
plt.imshow(imInput_bm[:,:,timePoint].T)

plt.plot([0,imInput_bm.shape[0]], [thetaSel, thetaSel], 'r')
plt.text(0.05 * imInput_bm.shape[0], thetaSel, 'Angle = %.3d' % (thetaSel+1), 
         fontsize=labelsize, color='r')
plt.gca().add_patch(mpatches.Rectangle((0, thetaSel-thetaWindow), imInput_bm.shape[0], 2*thetaWindow,
                                       fill=False, ec='g'))

plt.plot([minDepth,minDepth], [0, imInput_bm.shape[1]], 'm', linewidth=2)

plt.xlim([0,imInput_bm.shape[0]])
plt.ylim([0,imInput_bm.shape[1]])
plt.xlabel('Depth', fontsize=labelsize)
plt.ylabel('Angle', fontsize=labelsize)
plt.title('B-mode image at time point %d' %(timePoint+1), fontsize=20)
plt.gca().invert_yaxis()

# extract the data around the current scan line
imCurScanLinePatch = imInput_bm[:,thetaSel-thetaWindow:thetaSel+thetaWindow,:]

# display m-mode data at a select theta
plt.figure()
plt.imshow(imCurScanLinePatch.mean(1).T)
plt.xlabel('Depth', fontsize=labelsize)
plt.ylabel('Time', fontsize=labelsize)
plt.title('Mean M-mode image for Angle=%d +/- Window=%d' 
          % (thetaSel+1, thetaWindow), fontsize=20)

# create a matrix where each row represents the data of a scan line at a timepoint  
# each row will be treated as an observation for pca
imCurScanLinePatch = imCurScanLinePatch[minDepth:,:,:]
X = []
for tid in range(imInput_bm.shape[2]):
    if tid == 0:
        X = imCurScanLinePatch[:,:,tid].flatten()
    else:
        X = np.vstack((X, imCurScanLinePatch[:,:,tid].flatten()))
    
# perform pca on X
pca = sklearn.decomposition.PCA(n_components=X.shape[0])
pca.fit(X)

# plot percentage of variance explained
if showEigenVarianceCoverage:
    
    x = range(len(pca.components_))
    cumSumVar = np.cumsum(pca.explained_variance_ratio_)
    
    plt.figure()
    plt.plot(x, cumSumVar)
    plt.xlim([0,X.shape[0]])
    plt.yticks(np.arange(0.0,1.01,0.05))
    plt.xlabel('Number of principal components', fontsize=labelsize)
    plt.ylabel('Percentage of Variance Explained', fontsize=labelsize)       

# project each observation onto the top few eigen vectors
X_proj = pca.transform(X)

# display projections onto top eigen vectors and 
plt.figure(figsize=(30, 7.5))
for eid in range(numEigenVectors):
    
    # display projection
    plt.subplot(numEigenVectors,1,eid+1)
    plt.plot(1 + np.arange(X_proj.shape[0]), X_proj[:,eid], linewidth=4)
    plt.xlim([0,X_proj.shape[0]])
    plt.ylabel('%d' % (eid+1), fontsize=labelsize)    
    plt.yticks(())      

    if eid == 0:
        plt.title('Projections onto top %d eigen vectors for data at angle=%d' % (numEigenVectors, thetaSel+1),
                  fontsize=titlesize)
    
    if eid == numEigenVectors-1:
        plt.xlabel('Time', fontsize=labelsize)

# display autocorrelation plots of projections onto top eigen vectors
if False:
    
    for eid in range(numEigenVectors):
        
        # display projection
        plt.figure()
        pdplot.autocorrelation_plot(pd.Series(X_proj[:,eid]), linewidth=4)
        
        if eid == 0:
            plt.title('Autocorrelation of projection onto eigen vector %d for data at angle=%d' 
                      % (eid+1, thetaSel+1),
                      fontsize=titlesize)
        
        if eid == numEigenVectors-1:
            plt.xlabel('Lag', fontsize=labelsize)
```

## Look at projections onto the top eigen vector for a series of scan lines

```python
# set parameters
titlesize = 24
labelsize = 20

angleWindow = 10
angleList = np.linspace(angleWindow, imInput_bm.shape[1]-angleWindow, 20)
minDepth = 600

# perform eigen analysis on all theta
plt.figure(figsize=(30,40))

for aid in range(len(angleList)):
        
    curAngle = np.round(angleList[aid]).astype('uint8')    

    # display scan line on b-mode MIP
    plt.subplot(len(angleList)*2, 1, aid*2+1)
    
    plt.imshow(imInput_bm.max(2).T)
    
    plt.plot([0,imInput_bm.shape[0]], [curAngle, curAngle], 'r')
    plt.text(0.05 * imInput_bm.shape[0], curAngle, 'Angle = %.3d' % (curAngle+1), 
             fontsize=labelsize, color='r')
    plt.gca().add_patch(mpatches.Rectangle((0, curAngle-angleWindow), imInput_bm.shape[0], 2*angleWindow,
                                           fill=False, ec='g'))

    plt.plot([minDepth,minDepth], [0, imInput_bm.shape[1]], 'm', linewidth=2)
    
    plt.xlim([0,imInput_bm.shape[0]])
    plt.ylim([0,imInput_bm.shape[1]])
    #plt.xlabel('Depth', fontsize=labelsize)
    #plt.ylabel('Angle', fontsize=labelsize)    
    plt.gca().invert_yaxis()

    if aid == 0:
        plt.title('MIP of b-mode stack along time', fontsize=20)
        
    # extract the data around the current scan line
    imCurScanLinePatch = imInput_bm[minDepth:,curAngle-angleWindow:curAngle+angleWindow,:]

    # create a matrix where each row represents the data of a scan line at a timepoint  
    # each row will be treated as an observation for pca
    X = []
    for tid in range(imInput_bm.shape[2]):
        if tid == 0:
            X = imCurScanLinePatch[:,:,tid].flatten()
        else:
            X = np.vstack((X, imCurScanLinePatch[:,:,tid].flatten()))
    
    # perform pca on X
    pca = sklearn.decomposition.PCA(n_components=1)
    pca.fit(X)

    # project each observation onto the eigen vectors
    X_proj = pca.transform(X)

    # display projections onto top eigen vector
    plt.subplot(len(angleList)*2, 1, aid*2+2)
    
    plt.plot(1 + np.arange(X_proj.shape[0]), X_proj[:,0], linewidth=4)
    plt.xlim([0,X_proj.shape[0]])

    if aid == 0:
        plt.title('Projections onto top eigen vector', fontsize=titlesize)    
        
    if aid == len(angleList) - 1:        
        plt.xlabel('Time', fontsize=labelsize)
```
